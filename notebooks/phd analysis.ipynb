```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
```

```python
# Load the data
data = pd.read_csv(r"C:\Users\dolap\Downloads\main_model_class_balance_v3.csv")
data.head()
```

```python
df = data

```

```python
# Bar plot for class distribution
plt.figure(figsize=(12, 6))
sns.barplot(x='Class Name', y='Total Count', data=data)
plt.xticks(rotation=45, ha='right')
plt.title('Class Distribution')
plt.show()
```

```python
# Stacked bar plot for data splits
data.set_index('Class Name')[['Training Count', 'Validation Count', 'Test Count']].plot(
    kind='bar', stacked=True, figsize=(12, 6))
plt.title('Data Splits by Class')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Horizontal stacked bar plot for data splits
ax = data.set_index('Class Name')[['Training Count', 'Validation Count', 'Test Count']].plot(
    kind='barh',
    stacked=True,
    figsize=(12, 4),
    width=0.8,
    edgecolor='white')

# Add title and labels
plt.title('Data Splits by Class',
          fontsize=16,
          fontweight='bold')
plt.xlabel('Count',
           fontsize=12)
plt.ylabel('Class Name',
           fontsize=12)

# Customize y-axis tick labels to be italic
ax.set_yticklabels(
    ax.get_yticklabels(),
    fontstyle='italic',
    rotation=15)

# Adjust legend position to below the plot in a single column
plt.legend(
    fontsize=10,
    title_fontsize=12,
    loc='upper center',
    bbox_to_anchor=(0.5, -0.25),  # Position below the plot
    ncol=3  # Single column
)

# Show the plot
plt.tight_layout()
plt.show()
```
